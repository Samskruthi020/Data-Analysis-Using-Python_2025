```python
import pandas as pd
import numpy as np
# Load the dataset
file_path = '/content/insurance.csv'
dataset = pd.read_csv(file_path)

# 1. Vectors: Extract individual columns
age_vector = dataset['age'].values
sex_vector = dataset['sex'].values
expenses_vector = dataset['expenses'].values

# 2. Factors: Encode 'sex' and 'smoker' as categorical data (factors)
sex_factor = pd.Categorical(dataset['sex'])
smoker_factor = pd.Categorical(dataset['smoker'])

# 3. Lists: Combine different data types into a list
custom_list = [age_vector, sex_vector, smoker_factor, "Insurance Dataset"]

# 4. Matrices: Create a matrix from numerical columns
numerical_data = dataset[['age', 'bmi', 'children', 'expenses']].values
numerical_matrix = np.array(numerical_data)

# 5. Data Frames: Create a new data frame with selected columns
new_data_frame = dataset[['age', 'sex', 'expenses']].copy()

# Output results
print("Vector (age):", age_vector[:5])
print("Factor (sex):", sex_factor.categories.tolist())
print("List :", custom_list[:2])  # Show a subset for clarity
print("Matrix Shape:", numerical_matrix.shape)
print("New Data Frame Example:")
print(new_data_frame.head())
```

```python
import matplotlib.pyplot as plt

# Generate a scatter plot of age vs. expenses
plt.figure(figsize=(10, 6))
plt.scatter(dataset['age'], dataset['expenses'], c='blue', marker='+', alpha=0.7)

# Customize the plot
plt.title('Scatter Plot of Age vs. Expenses', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Expenses ($)', fontsize=12)
plt.grid(alpha=0.3)

# Add text annotations for specific points (e.g., highest expense)
max_expense_idx = dataset['expenses'].idxmax()
plt.text(dataset['age'][max_expense_idx], dataset['expenses'][max_expense_idx] + 2000,
         'Highest Expense', color='red', fontsize=10, ha='center')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
file_path = '/content/insurance.csv'  # Update this path to your file location
dataset = pd.read_csv(file_path)
# Filter the data: Select individuals who are smokers
filtered_data = dataset[dataset['smoker'] == 'yes']
# Generate a scatter plot for the filtered data (smokers only)
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['age'], filtered_data['expenses'], c='red', marker='x', alpha=0.7)
# Customize the plot
plt.title('Scatter Plot of Age vs. Expenses (Smokers Only)', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Expenses ($)', fontsize=12)
plt.grid(alpha=0.3)
# Add a note about the filter
plt.text(18, filtered_data['expenses'].max() + 5000, 'Filtered: Smokers Only',
         fontsize=12, color='purple', ha='left', va='center')
# Show the plot
plt.show()
```
